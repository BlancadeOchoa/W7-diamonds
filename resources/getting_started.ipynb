```python
#We create the dummies for the columns we want the type to change
dmd_dummy = pd.get_dummies(dmd[['cut','color','clarity']], drop_first=True)
dmd_dummy.head()
```

```python
#I now drop the columns I dont need
dmd_2=dmd.drop(columns=["cut","color","clarity"])
```

```python
#I now join the dummies to the rest of the table
dmd_2=dmd_2.join(dmd_dummy)
dmd_2.head()
```

```python
dmd_2.dtypes
```

## Starting with a regression model

```python
X = dmd_2.drop(columns='price')
y = dmd_2['price'] 
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
#Separate those into test and training data with 80% of the data in the training set
X_train, X_test, y_train, y_test = train_test_split( X, y, train_size=0.8)
```

```python
model1 = LinearRegression()
model1.fit(X_train, y_train)
```

```python
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error
```

### r2 score

```python
#Training set
y_pred_train = model1.predict(X_train)

r2_ = r2_score(y_pred=y_pred_train, y_true=y_train)
r2_
```

```python
#Testing set
y_pred_test = model1.predict(X_test)

r2_test = r2_score(y_pred=y_pred_test, y_true=y_test)
r2_test
```

### mse

```python
#Training set
mse_ = mean_squared_error(y_true=y_train, y_pred= y_pred_train) 
mse_
```

```python
#Testing set
mse_test = mean_squared_error(y_true= y_test, y_pred= y_pred_test) 
mse_test
```

squaredboolean value, optional (default = True)

    If True returns MSE value, if False returns RMSE value.


As a default its true, and we want it to be rmse not mse, therefore we change it to False

### rmse 

```python
rmse_ = mean_squared_error(y_true= y_train, y_pred= y_pred_train, squared=False) #if False returns RMSE value.
rmse_
```

```python
rmse_test = mean_squared_error(y_true= y_test, y_pred= y_pred_test, squared=False)
rmse_test
```
