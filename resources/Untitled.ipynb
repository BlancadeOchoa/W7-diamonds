```python
import pandas as pd
```

```python
dmd_shot4 = pd.read_csv("train.csv", index_col="id") 
```

```python
dmd_shot4.columns
```

```python
dmd_shot4 = dmd_shot4.drop(columns=['x','y','z']).head()
```

```python
dmd_shot4.head()
```

```python
dmd_shot4['cut'].unique
```

```python
#We create the dummies for the columns we want the type to change
dmd_dummy = pd.get_dummies(dmd_shot4[['cut','color','clarity']], drop_first=True)
dmd_dummy.head()
```

```python
#I now drop the columns I dont need
dmd_2=dmd_shot4.drop(columns=["cut","color","clarity"])
dmd_2
```

```python
#I now join the dummies to the rest of the table
dmd_2=dmd_2.join(dmd_dummy)
dmd_2.head()
```
