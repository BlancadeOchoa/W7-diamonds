# 4. DATA CLEANING:
- Data types : replace categories whith numbers (manually)
- Drop columns : low colinearity



```python
#Import libraries:
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#import the dataset:
%store -r dmd
```

```python
dmd.head()
```

## Data types 

```python
dmd["cut"].replace({"Ideal": 5, "Premium": 4, "Very Good" : 3, "Good" : 2, "Fair" : 1}, inplace=True)
dmd["color"].replace({"D": 7, "E": 6, "F" : 5, "G" : 4, "H" : 3, "I": 2 , "J": 1}, inplace=True)
dmd["clarity"].replace({"IF": 9, "VVS1": 8, "VVS2" : 7, 
                        "VS1" : 6, "VS2" : 5, "SI1": 4, 
                        "SI2" : 2, "I1": 1 }, inplace=True)
```

```python
dmd.head()
```

## Drop columns

### We will check the colinearity again and look for a different approach

```python
plt.figure(figsize=(15, 12))

#sns.heatmap(websites.corr()); #This is the heatmap withput numbers
heatmap = sns.heatmap(dmd.corr(), vmin=-1, vmax=1, annot=True)
```

```python
#We will check columns whith very low correlation
vlow_corr = dmd.corr() > 0.10
vlow_corr
```

We can see how depth has a very low correlation whith price hence why we choose to drop it

```python
drop1_lc = dmd.drop(columns=['depth'])
```

```python
#Lets check again for other low correlations
low_corr = drop1_lc.corr() > 0.20
low_corr
```

'table', 'cut', 'color' and 'clarity' have very low correlations as well, but we know that the 4c are crucial when pricing a diamond (https://selectingadiamond.com/diamond-pricing-a-no-nonsense-guide-to-the-4-cs/), hence why we choose to modify them in order to keep them

#### In the colinearity heatmap we can see how carat has a strong colinearity with price, hence why I choose to use cut, color and clarity and divide it by carat (weight):

```python
drop1_lc['cut_weight']=drop1_lc['cut']/drop1_lc['carat']
drop1_lc['color_weight']=drop1_lc['color']/drop1_lc['carat']
drop1_lc['clarity_weight']=drop1_lc['clarity']/drop1_lc['carat']
dmd_cl4 = drop1_lc.drop(['cut','color','clarity','table'], axis=1)
```
